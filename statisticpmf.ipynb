## PMF
What is the probability of getting exactly 5 heads when flipping a fair coin 12 flips?

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

# Coin flip scenario
n = 12
p = 0.5

x =np.arange(0, n+1)
pmf_vals = binom.pmf(x, n, p)

plt.stem(x, pmf_vals)
plt.title("Binomial PMF Prob of Heads in 12 coin flips")
plt.xlabel("No. of Heads")
plt.ylabel("Probability")
plt.show()
           
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

# Coin flip scenario
n = 12
p = 0.5

x =np.arange(0, n+1)

pmf_heads = binom.pmf(x, n, p)
pmf_tails = binom.pmf(x, n, 1-p)

plt.stem(x, pmf_heads, linefmt='b-', markerfmt='bo', basefmt=' ', label='Heads')
plt.stem(x, pmf_tails, linefmt='r-', markerfmt='ro', basefmt=' ', label='Tails')
plt.title("Binomial PMF Prob of Heads and Tails in 12 coin flips")

plt.xlabel("No. of Outcomes")
plt.ylabel("Probability")
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

# Coin flip scenario
n = 12
p = 0.5

x =np.arange(0, n+1)

pmf_heads = binom.pmf(x, n, p)    # Using p for heads, flip coin x times
pmf_tails = binom.pmf(n-x, n, p)  # Using n-x to represent tails, flip coin x times

width = 0.4
plt.bar(x - width/2, pmf_heads, color='b', label='Heads')
plt.bar(x + width/2, pmf_tails, color='g', label='Tails')
plt.title("Binomial PMF Prob of Heads and Tails in 12 coin flips")
plt.xlabel("No. of Outcomes")
plt.ylabel("Probability")
plt.legend()
plt.show()
```

What is the probability that exactly 4 customers arrive in a minute of the average is 5 per minute?

```python
import numpy as np
from scipy.stats import poisson
import matplotlib.pyplot as plt

# parameters
lam = 5
x = np.arange(0, 5)
pmf_vals = poisson.pmf(x, lam)

# plot all bars in blue color
markerline, stemlines, baseline = plt.stem(x, pmf_vals)
plt.setp(stemlines, color='b', linewidth=1)
plt.setp(markerline, color='b', markersize=6)

# Highlight k=4 in red
k = 4
plt.stem([k], [pmf_vals[k]], linefmt='r', markerfmt='ro', basefmt=' ')

plt.title("Poisson PMF - Customers per minute (=5) Red Highlight = P(X=4)")
plt.xlabel("No. of Customers")
plt.ylabel("Probability")
plt.show()

```

## PDF - Continuous 
### Exponential distribution - used for timebound case
### Normal distribution - has mean and standard distribution


What is the likelihood that a randomly selected student height is between 160cm and 170cm

```python
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt

# parameters: assume mean height 165cm , std dev 7cm
mu = 165
sigma = 7

# Continuous range of heights
x = np.linspace(140, 190, 400)    # range of heights from 140cm to 190cm
pdf_vals = norm.pdf(x, mu, sigma) # Calculate PDF values

# Probability for height between 160cm and 170cm
lower, upper = 160, 170
prob_range = norm.cdf(upper, mu, sigma) -norm.cdf(lower, mu, sigma)

# Plotting the PDF
plt.plot(x, pdf_vals, label='PDF')
plt.fill_between(x, pdf_vals, where=(x>=lower) & (x<=upper), color="skyblue", alpha=0.5, label=f'P({lower}cm  X  {upper}cm)={prob_range:.3f}')
plt.title("Normal Distribution of Heights (=165cm, =7cm)")
plt.xlabel("Height (cm)")
plt.ylabel("Probability Density")
plt.legend()
plt.show()

```

### What is the probability that a customer waits between 2 and 5 minutes?

```python
from scipy.stats import expon

# Average waiting time = 3.5 minutes
scale = 3.5 # mean = scale

x = np.linspace(0, 10, 400)   # Range from 0 to 10 minutes, 400 points 
pdf_vals = expon.pdf(x, scale=scale)  # Calculate PDF values

# probability of waiting between 2 and 5 minutes
lower, upper = 2, 5
prob_range = expon.cdf(upper, scale=scale) - expon.cdf(lower, scale=scale)

# plot PDF
plt.plot(x, pdf_vals, label='PDF')
plt.fill_between(x, pdf_vals, where=(x>=lower) & (x<=upper), color="lightgreen", alpha=0.5, label=f'P({lower}  X  {upper})={prob_range:.3f}')
plt.title("Exponential Distribution of Waiting Times (mean=3.5 mins)")
plt.xlabel("Waiting Time (minutes)")
plt.ylabel("Probability Density")
plt.legend()
plt.show()
```
