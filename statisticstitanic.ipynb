Probability Mass Function (PMF)

```python
# Binomial Distribution
# 4 People playing ludo. Simulating a dice roll

np.random.seed(42)                            # random seed for reproducibility
rolls = np.random.randint(1, 7, size=4)        # randint generates random integers from low (inclusive) to high (exclusive)

faces = np.arange(1, 7)  # faces, frequency of the appearance for the dice 1/6.2/6 etc
counts = [np.sum(rolls == face) for face in faces]

pmf_empirical = counts/sum(counts) # calling the binomial function using pmf_empirical = counts/sum(counts)
```

```python
# plotting the PMF
x = [1,2,3,4,5,6]
plt.stem(faces,pmf_empirical)
plt.xlabel('Dice_outcomes')
plt.ylabel('Likelihood')
plt.title('Four Pairs Simulation')
plt.ylim(0, 1)
plt.show()
```

```python
#class work
# The Area under the curve represent the total
# plot an histogram
# Convert age into normal distribution - Mean, Standard Deviation

from matplotlib.pylab import norm
from scipy.stats import norm

normal_ages = df['Age']
mu = normal_ages.mean()
sigma = normal_ages.std()
# plot the normal distribution curve
x = np.linspace(normal_ages.min(), normal_ages.max(), 100)

ages_pdfs = norm.pdf(x, mu, sigma)

plt.figure(figsize=(10,6))
plt.hist(normal_ages, bins=30, density=True, alpha=0.6, color='g')
plt.plot(x, ages_pdfs, 'k', linewidth=2)
plt.xlabel('Age')
plt.ylabel('Density')
plt.title('Age Distribution on the Titanic with Normal Distribution Fit, Age_Pdf')
plt.show()

```

```python
# Alternaive

normal_ages = df['Age']
mu = normal_ages.mean()
sigma = normal_ages.std()
# plot the normal distribution curve
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 100)
pdf = stats.norm.pdf(x, mu, sigma)
plt.plot(x, pdf, color='blue')
plt.xlabel('Age')
plt.ylabel('Density')
plt.title('Normal Distribution of Ages on the Titanic')
plt.show()
```

```python
# Alternative - Normal Distribution

# The Area under the curve represent the total
# plot an histogram
# Convert age into normal distribution - Mean, Standard Deviation
age_mean = df['Age'].mean()
age_std = df['Age'].std()
age_range = np.linspace(0, 80, 100)
age_pdf = stats.norm.pdf(age_range, age_mean, age_std)
plt.plot(age_range, age_pdf, color='blue')
plt.fill_between(age_range, age_pdf, color='blue', alpha=0.5)
plt.title('Normal Distribution of Ages on the Titanic')
plt.xlabel('Age')
plt.ylabel('Density')
plt.show()  
```
