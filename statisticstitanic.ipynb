Probability Mass Function (PMF)

```python
# Binomial Distribution
# 4 People playing ludo. Simulating a dice roll

np.random.seed(42)                            # random seed for reproducibility
rolls = np.random.randint(1, 7, size=4)        # randint generates random integers from low (inclusive) to high (exclusive)

faces = np.arange(1, 7)  # faces, frequency of the appearance for the dice 1/6.2/6 etc
counts = [np.sum(rolls == face) for face in faces]

pmf_empirical = counts/sum(counts) # calling the binomial function using pmf_empirical = counts/sum(counts)
```

```python
# plotting the PMF
x = [1,2,3,4,5,6]
plt.stem(faces,pmf_empirical)
plt.xlabel('Dice_outcomes')
plt.ylabel('Likelihood')
plt.title('Four Pairs Simulation')
plt.ylim(0, 1)
plt.show()
```
